**(3) Photo-voltaic generation converts the sun's radiation into usable electricity. In this task, you will get the *direct solar radiation* with 1 minute time interval from the *Radiation observations*  at [FMI](http://en.ilmatieteenlaitos.fi/download-observations#!/). Plot radiation profile of three different days so that one must be in March, other in July and the last in December (regardless of the year). Provide *information* about the potential of solar generation in those days. You can also select the measuring station (but write it in the answer).**

*Hint:* Code like in the tutorial notebook.

```python
#Import libraries
import pandas as pd #https://pandas.pydata.org/
import matplotlib.pyplot as plt #https://matplotlib.org/
import matplotlib.dates as mdates #https://matplotlib.org/
```

### Importing all solar radiation data from Helsinki Kumpula

```python
march_radiation = pd.read_csv("march-solar-radiation.csv", 
                                sep=",",
                                dayfirst=True,
                                header=0,
                                decimal=b".",
                                index_col=0, 
                                parse_dates= [[0, 1, 2, 3]],
                                usecols=[0,1,2,3,5])
march_radiation.info()
march_radiation.head()
```

```python
july_radiation = pd.read_csv("july-solar-radiation.csv", 
                                sep=",",
                                dayfirst=True,
                                header=0,
                                decimal=b".",
                                index_col=0, 
                                parse_dates= [[0, 1, 2, 3]],
                                usecols=[0,1,2,3,5])
july_radiation.info()
july_radiation.head()
```

```python
december_radiation = pd.read_csv("december-solar-radiation.csv", 
                                sep=",",
                                dayfirst=True,
                                header=0,
                                decimal=b".",
                                index_col=0, 
                                parse_dates= [[0, 1, 2, 3]],
                                usecols=[0,1,2,3,5])
december_radiation.info()
december_radiation.head()
```

### Plotting each datasets

```python
# march 2020
fig, ax = plt.subplots(figsize=(16,6))
ax.plot(march_radiation, color='blue', linestyle='--')
ax.set_title("20 March 2020")
ax.grid(True)
ax.set_ylabel("Direct solar radiation (W/m2)")

xfmt = mdates.DateFormatter('%H:%M')
ax.xaxis.set_major_formatter(xfmt)
plt.show()
```

```python
# July 2020
fig, ax = plt.subplots(figsize=(16,6))
ax.plot(july_radiation, color='black', linestyle='--')
ax.set_title("17 July 2020")
ax.grid(True)
ax.set_ylabel("Direct solar radiation (W/m2)")

xfmt = mdates.DateFormatter('%H:%M')
ax.xaxis.set_major_formatter(xfmt)
plt.show()
```

```python
# december 2020
fig, ax = plt.subplots(figsize=(16,6))
ax.plot(december_radiation, color='green', linestyle='-')
ax.set_title("18 December 2020")
ax.grid(True)
ax.set_ylabel("Direct solar radiation (W/m2)")

xfmt = mdates.DateFormatter('%H:%M')
ax.xaxis.set_major_formatter(xfmt)
plt.show()
```

### Plotting three different days in one single plot

```python
#Difference
d1 = march_radiation.index[0].date() - july_radiation.index[0].date()
#
d2 = march_radiation.index[0].date() - december_radiation.index[0].date()

fig, ax = plt.subplots(1,figsize=(16,6))
#
plt.plot(march_radiation.index, march_radiation.values, label="20 March 2020", color='green',linestyle='--')
#
plt.plot(july_radiation.index + pd.offsets.Day(d1.days), 
         july_radiation.values, label="17 July 2020", color='blue',linestyle='-')
#
plt.plot(december_radiation.index + pd.offsets.Day(d2.days), 
         december_radiation.values, label="18 December 2020", color='red',linestyle=':')
#
plt.legend()
plt.title("Direct Solar Radiation") 
plt.ylabel("Direct solar radiation [W/m2]") 
plt.grid(True)
#change x ticks
xfmt = mdates.DateFormatter('%H:%M')
ax.xaxis.set_major_formatter(xfmt)
#
plt.show()
```

Direct solar radiation is measured in the unit of Watt per square meter (W/m2). As we know, watt is the unit of power and through this measurement, we can calculate the potential solar energy in the area of measurement. According to the Department of Energy of the United States' website (https://www.energy.gov/eere/solar/solar-photovoltaic-technology-basics), one solar PV cell typically can produce around 1-2 Watts of power. Of course, there are other external conditions that would affect the energy production efficiency, but using this correlation we can roughly measure the potential of solar energy in particular area.

From these knowledge, we can summarize that the potential power production of solar energy varies between seasons and days of the year. In the summer (17 July 2020, in this case), the sun radiates light and radiation for the longest time period, which increases the amount of power that theoretically can be produced by a solar PV cell. In the winter, we can see that the solar PV do not have much potential in terms of energy generation. Based on empirical evidence (experience that we have living in Finland), the graph above illustrates winter conditions in Finland very well, which is mostly cloudy/dark with minimal sun presence or just occasional sunny days. It also means that the solar PV is virtually almost useless in the winter. In the spring (20 March 2020), we can see the pattern that we could also see in countries close to the equator, in which the duration of the solar radiation might not be as long as in the summer but the amount of it in certain times could even surpass the measurement taken in the summer. This also makes solar PV viable to use as energy generation.
