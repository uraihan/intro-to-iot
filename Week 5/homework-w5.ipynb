**(1) Create two arrays with 10000 samples generated as [Binomial](https://en.wikipedia.org/wiki/Binomial_distribution) random variables via [Numpy](https://docs.scipy.org/doc/numpy/reference/generated/numpy.random.binomial.html) similar to what we have done with Poisson random variables in the tutorial considering the following paramenters of the Binomial function:**

**(a) $n=100$ and $p=0.3$;**

**(b) $n=10000$, and $p=0.6$**


**Make the histogram (empirical) plots related to each vector and compare them with the analytical results with the probability distribution funtion given by:**

**$$\mathrm{Prob}\left[X = k \right] = \binom{n}{k}p^k(1-p)^{n-k},$$ where $k$ is the number of *successes* in $n$ trials.**

**Remember to use this [scipy.stats.binom](https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.stats.binom.html) in a similar way we did in the tutorial with Poisson.**

```python
import numpy as np 
import matplotlib.pyplot as plt
from matplotlib import style
from scipy.stats import binom
style.use('bmh')
#Not to show warning messages (to keep the notebook clean)
import warnings
warnings.filterwarnings('ignore')
```

### Case (a)

```python
n, p = 100, 0.3
a = np.random.binomial(n,p,10000)
n_bins3 = np.max(a) - np.min(a)
plt.figure(figsize=(14,5))
plt.hist(a,n_bins3,density=True)
plt.xlabel('$n$')
plt.ylabel('Frequency of $N = n$')
plt.title('Binomial distribution')
plt.show()
```

#### Analytical Distribution from mathematical formula

```python
n = np.arange(50)
plt.figure(figsize=(14,5))
plt.step(n, binom.pmf(n, 100, 0.3), where='pre', label='$n=100, p=0.3$')
plt.xlim(left=10)
```

### Case (b)

```python
n, p = 10000, 0.6
b = np.random.binomial(n,p,10000)
n_bins6 = np.max(b) - np.min(b)
plt.figure(figsize=(14,5))
plt.hist(b,n_bins6,density=True)
plt.xlabel('$n$')
plt.ylabel('Frequency of $N = n$')
plt.title('Binomial distribution')
plt.show()
```

#### Analytical Distribution from mathematical formula

```python
n = np.arange(6200)
plt.figure(figsize=(14,5))
plt.step(n, binom.pmf(n, 10000, 0.6), where='pre', label='$n=10000, p=0.6$')
plt.xlim([5500, 6200])
```

```python
plt.figure(figsize=(14,5))
plt.hist(a,n_bins3,density=True)
plt.hist(b,n_bins6,density=True)
plt.xlabel('$n$')
plt.ylabel('Frequency of $N = n$')
plt.title('Binomial distribution')
plt.show()
```

```python
n = np.arange(6200)
plt.figure(figsize=(14,5))
plt.step(n, binom.pmf(n, 100, 0.3), where='pre', label='$n=100, p=0.3$')
plt.step(n, binom.pmf(n, 10000, 0.6), where='pre', label='$n=10000, p=0.6$')
plt.xlabel('$n$')
plt.ylabel('Probability that $N = n$')
plt.title('Binomial distribution')
#plt.ylim([0, 1])
plt.grid(True) #grid
plt.legend()
plt.show() 
```

**(2) Read Section 2.3 and Chapter 3 from [Network Science](http://networksciencebook.com/) and generate three different kinds of Erdos-Renyi graphs with $N=15$ (fifteen nodes) using [NetworkX](https://networkx.github.io/documentation/stable/reference/generated/networkx.generators.random_graphs.erdos_renyi_graph.html?highlight=erdos#networkx.generators.random_graphs.erdos_renyi_graph). 
The networks shall be (a)  with probability $p=0.2$, (b) with probability $p=0.5$ and (c) with probability $p=0.8$. Discuss about the differences you see if these graphs represent communication networks.**

```python
import networkx as nx #https://networkx.github.io/
```

```python
A = nx.erdos_renyi_graph(15, 0.2)
B = nx.erdos_renyi_graph(15, 0.5)
C = nx.erdos_renyi_graph(15, 0.8)

plt.figure(figsize=(14,5))
ax1 = plt.subplot(1,3,1)
ax1.set_title("Network A")
nx.draw_networkx(A, node_color='black', font_color='white', edge_color='black', font_weight='bold',  node_size=700)

ax2 = plt.subplot(1,3,2)
ax2.set_title("Network B")
nx.draw_networkx(B, node_color='black', font_color='white', edge_color='black', font_weight='bold',  node_size=700)

ax3 = plt.subplot(1,3,3)
ax3.set_title("Network C")
nx.draw_networkx(C, node_color='black', font_color='white', edge_color='black', font_weight='bold',  node_size=700)

plt.show()
```

From these three networks, the most visible difference is the connectivity of each nodes with another nodes. In Network A, the connection between nodes looks a bit loose even though they are all connected with at least one or more other nodes. The higher the value we put for the probability argument, the higher connection each nodes will make with other nodes. This is visible if we look at Network C, where each nodes are highly connected with other nodes. The variation in probability and its effect on the network can be observed from Network A where in some cases, the random generated graph could leave at least one node disconnected with the rest of the network.

In the context of communication networks, Network A represent a loosely communicated communicated network, Network C represents a highly connected communication network, and Network B represent something in between Network A and C. Transfer of data in Network A might take more time to travel from one node to another, while in Network B and C transfer of data is more likely to be faster because each node is connected to more nodes thus having more efficient path route to transfer data to another nodes.
