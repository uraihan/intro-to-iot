**(2) Assume that graph illustrate a scenario of four sensor nodes that also (pre-)process data and one aggregator node as follows. Nodes $B$ and $C$ monitor temperature and wind speed at Airport station in Lappeenranta, respectively. Nodes $D$ and $E$ monitor temperature and wind speed at Lepola station in Lappeenranta, respectively. Nodes $B$ and $D$ send a binary message to $A$ indicating if during that period the temperature was more than $25$ degrees ("1" means above, "0" means below or equal to). Likewise nodes $C$ and $E$ send a binary message to $A$ indicating if during that period the wind speed was more than $5$ m/s ("1" means above, "0" means below or equal to). Download and plot the data from [FMI](http://en.ilmatieteenlaitos.fi/download-observations\#!/) with 1 hour resolution during July, 2018. Using this data, plot the binary signals sent from the nodes $B$-$E$ to the aggregator $A$.**

```python
#Import libraries
import pandas as pd #https://pandas.pydata.org/
#import matplotlib.pyplot as plt #https://matplotlib.org/
import matplotlib.dates as mdates #https://matplotlib.org/
#import numpy as np #http://www.numpy.org/
#from datetime import date #https://docs.python.org/3/library/datetime.html
```

```python
# read csv data from Lappeenranta Lepola
lepola = pd.read_csv("lpr-lepola.csv", dayfirst=True, sep=",",
                      header=0, decimal=b".", index_col=0, 
                      parse_dates=[[0,1,2,3]], usecols=[0,1,2,3,5,6])

lepola.info()
```

```python
# read csv data from Lappeenranta Lentoasema
lentoasema = pd.read_csv("lpr-lentoasema.csv", dayfirst=True, sep=",",
                      header=0, decimal=b".", index_col=0, 
                      parse_dates=[[0,1,2,3]], usecols=[0,1,2,3,5,6])

lentoasema.info()
```

```python
# assigning each nodes
nodes_b = lentoasema.drop(columns="Wind speed (m/s)")
nodes_c = lentoasema.drop(columns="Air temperature (degC)")
nodes_d = lepola.drop(columns="Wind speed (m/s)")
nodes_e = lepola.drop(columns="Air temperature (degC)")

nodes_b.info();
nodes_c.info();
nodes_d.info();
nodes_e.info();
```

```python
# Plotting findings during July 2018
fig, ax1 = plt.subplots(figsize=(16,6))
ax1.set_ylabel("Air Temperature (degC)")
ax1.plot(nodes_b>25,color='red', label="Node B", marker='o',linestyle='-')

ax2 = ax1.twinx()
ax2.set_ylabel("Wind Speed (m/s)", color='black',rotation=270,va="bottom")
ax2.plot(nodes_c>5,color='blue', label="Node C", marker='x',linestyle='--')

ax3 = ax1.twinx()
ax3.plot(nodes_d>25,color='green', label="Node D", marker='s',linestyle='-')

ax4 = ax1.twinx()
ax4.plot(nodes_e>5,color='yellow', label="Node E", marker='1',linestyle='--')

ax1.grid(True)
plt.title("Binary signals sent to node A from nodes B, C, D, and E around Lappeenranta in July 2018")

xfmt = mdates.DateFormatter('%D')
ax1.xaxis.set_major_formatter(xfmt)

plt.show()
```

In the plot above, Red is for Node A, Blue for Node B, Green for Node C, and Yellow for Node E

**(3) Node $A$ needs to identify risky situations in Lappeenranta when the temperature is above $25$ degrees and the wind is above $5$ m/s based on the sensors' signals. If a risky situation is identified at node $A$ as binary number "1" while normal situations as "0". How would you solve this problem? Plot the output signal from node $A$ and discuss its relation to the input signals from $B$-$E$.**

```python
risky = ((nodes_b.values>25) & (nodes_c.values>5)) | ((nodes_d.values>25) & (nodes_e.values>5))

plt.figure(figsize=(16,6))
plt.plot(nodes_b.index, risky, color='green', marker='s',linestyle='-')

plt.grid(True)

plt.show()
```

The plot above is the output signal from node A identifying risky situations in Lappeenranta. The logic used here is the combination of AND and OR logic gates. Basically, when the temperature and wind speed in either one of the stations exceeded the threshold (25 degrees Celsius and 5 m/s, respectively), then node A will send a signal "1" identifying a risky situation. 

So in this case, I designed node A to detect a risky situation if either one of the stations detected a risky situation based on air temperature and wind speed. The reasoning behind this is, if we look at the previous plot in question (2), the air temperature and wind speed measurement in the Lepola station doesn't always exceed 25 degree Celsius and 5 m/s at the same time. So if we use the logic gates AND to connect Airport station (node B and C) and Lepola station (node D and E), node A won't be able to detect risky situation in any of the measurement station area, as shown in graph below.

```python
risky2 = ((nodes_b.values>25) & (nodes_c.values>5)) & ((nodes_d.values>25) & (nodes_e.values>5))

plt.figure(figsize=(16,6))
plt.plot(nodes_b.index, risky2, color='red', marker='s',linestyle='-')

plt.grid(True)

plt.show()
```
